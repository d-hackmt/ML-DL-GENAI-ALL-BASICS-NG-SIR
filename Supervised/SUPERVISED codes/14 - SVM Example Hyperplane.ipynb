```python
import numpy as np
import matplotlib.pyplot as plt

# Generating synthetic data for two classes
np.random.seed(0)
class_1 = np.random.randn(20, 2) + [2, 2]  # Class 1 centered at (2, 2)
class_2 = np.random.randn(20, 2) + [-2, -2]  # Class 2 centered at (-2, -2)

# Plotting the data
plt.scatter(class_1[:, 0], class_1[:, 1], marker='o', label='Class 1')
plt.scatter(class_2[:, 0], class_2[:, 1], marker='x', label='Class 2')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Synthetic Data for Binary Classification')
plt.grid(True)
plt.show()
```

```python
from sklearn.svm import SVC

# Concatenating the data and labels
X = np.vstack((class_1, class_2))
y = np.hstack((np.ones(20), -np.ones(20)))  # Class 1: 1, Class 2: -1

# Training a linear SVM model
svm_model = SVC(kernel='linear')
svm_model.fit(X, y)

# Plotting decision boundary and support vectors
plt.scatter(class_1[:, 0], class_1[:, 1], marker='o', label='Class 1')
plt.scatter(class_2[:, 0], class_2[:, 1], marker='x', label='Class 2')

# Plotting decision boundary
w = svm_model.coef_[0]
b = svm_model.intercept_[0]
slope = -w[0] / w[1]
xx = np.linspace(-4, 4)
yy = slope * xx - (b / w[1])
plt.plot(xx, yy, 'k-', label='Decision Boundary')

# Plotting support vectors
support_vectors = svm_model.support_vectors_
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=100,
            facecolors='none', edgecolors='k', label='Support Vectors')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Binary Classification with SVM')
plt.grid(True)
plt.show()
```
