```python
import numpy as np
from scipy.stats import skew, kurtosis

# Generate random data for illustration
data_1 = np.random.normal(0, 1, 1000)  # Standard normal distribution
data_2 = np.random.uniform(-1, 1, 1000)  # Uniform distribution


# Calculate skewness and kurtosis for each dataset
skewness_1, kurtosis_1 = skew(data_1), kurtosis(data_1)
skewness_2, kurtosis_2 = skew(data_2), kurtosis(data_2)

# Print the results
print("Dataset 1 - Standard Normal Distribution:")
print("Skewness:", skewness_1)
print("Kurtosis:", kurtosis_1)
print()

print("Dataset 2 - Uniform Distribution:")
print("Skewness:", skewness_2)
print("Kurtosis:", kurtosis_2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, norm
from scipy import stats

# Generate random data for illustration
data_1 = np.random.normal(0, 1, 1000)  # Standard normal distribution
data_2 = np.random.uniform(-1, 1, 1000)  # Uniform distribution

# Function to plot histogram and fitted normal distribution
def plot_skew_kurt(data, title):
    plt.figure(figsize=(12, 6))

    # Plot histogram
    sns.histplot(data, kde=False, bins=30, color='blue', label='Histogram')

    # Fit a normal distribution to the data
    mu, std = stats.norm.fit(data)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2, label='Fitted Normal Distribution')

    plt.title(title)
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

plot_skew_kurt(data_1, 'Dataset 1 - Standard Normal Distribution')

plot_skew_kurt(data_2, 'Dataset 2 - Uniform Distribution')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, norm
from scipy import stats

# Example 1: Positively skewed data
data_pos_skew = np.random.exponential(scale=2, size=1000)

# Example 2: Negatively skewed data
data_neg_skew = -np.random.exponential(scale=2, size=1000)

# Function to plot histogram and fitted normal distribution
def plot_skew_kurt(data, title):
    plt.figure(figsize=(12, 6))

    # Plot histogram
    sns.histplot(data, kde=False, bins=30, color='blue', label='Histogram')

    # Fit a normal distribution to the data
    mu, std = stats.norm.fit(data)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2, label='Fitted Normal Distribution')

    plt.title(title)
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

# Plot for Example 1
plot_skew_kurt(data_pos_skew, 'Example 1 - Positively Skewed Data')

# Plot for Example 2
plot_skew_kurt(data_neg_skew, 'Example 2 - Negatively Skewed Data')
```

```python
import numpy as np
from scipy.stats import pearsonr

# Example data
variable1 = np.array([1, 2, 3, 4, 5])
variable2 = np.array([2, 4, 5, 4, 5])

# Calculate Pearson correlation coefficient and p-value
corr_coefficient, p_value = pearsonr(variable1, variable2)

# Display the results
print(f"Pearson Correlation Coefficient: {corr_coefficient}")
print(f"P-value: {p_value}")

# Interpret the correlation
if corr_coefficient > 0:
    print("There is a positive correlation.")
elif corr_coefficient < 0:
    print("There is a negative correlation.")
else:
    print("There is no correlation between the variables.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Example data
variable1 = np.array([1, 2, 3, 4, 5])
variable2 = np.array([2, 4, 5, 4, 5])

# Calculate Pearson correlation coefficient and p-value
corr_coefficient, _ = pearsonr(variable1, variable2)

# Plot the scatter plot
plt.scatter(variable1, variable2, label=f'Correlation: {corr_coefficient:.2f}', color='blue')

# Plot the line of best fit
plt.plot(np.unique(variable1), np.poly1d(np.polyfit(variable1, variable2, 1))(np.unique(variable1)),
         color='red', linestyle='--', label='Line of Best Fit')

# Add labels and title
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Scatter Plot with Line of Best Fit')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Example data
data = {'Age': [10, 1, 2, 2, 2, 3, 0, 3, 3, 4, 5, 5, 6, 6]}
df = pd.DataFrame(data)

# Step 1: Count the number of observations
n = len(df)
print(n)

# Step 2: Sort the observations
#sorted_data = np.sort(df['Age'])

# Step 3: Calculate the quartiles
Q1 = np.percentile(df, 25)
Q2 = np.percentile(df, 50)
Q3 = np.percentile(df, 75)

# Step 4: Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Step 5: Calculate the lower and upper bounds for identifying outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 6: Visualize quartiles with a boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'], vert=False)
plt.title('Boxplot of Ages with Quartiles')
plt.xlabel('Age')
plt.show()

# Step 7: Print quartile information
print(f'Q1 (25th percentile): {Q1}')
print(f'Q2 (50th percentile): {Q2}')
print(f'Q3 (75th percentile): {Q3}')
print(f'IQR (Interquartile Range): {IQR}')
print(f'Lower Bound for Outliers: {lower_bound}')
print(f'Upper Bound for Outliers: {upper_bound}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Example data
data = {'Score': [65, 72, 78, 82, 88, 90, 92, 95, 98, 100]}
df = pd.DataFrame(data)

# Calculate percentiles
median = np.percentile(df['Score'], 50)
q25 = np.percentile(df['Score'], 25)
q75 = np.percentile(df['Score'], 75)
percentiles = [q25, median, q75]

# Calculate deciles
deciles = [np.percentile(df['Score'], i * 10) for i in range(1, 10)]

# Visualize with a boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(df['Score'], vert=False)
plt.title('Boxplot with Percentiles and Deciles')
plt.xlabel('Score')
plt.show()

# Print percentile and decile information
print('Percentiles:')
print(f'25th Percentile (Q1): {q25}')
print(f'50th Percentile (Median): {median}')
print(f'75th Percentile (Q3): {q75}')

print('\nDeciles:')
for i, decile in enumerate(deciles, start=1):
    print(f'D{i} ({i * 10}th Percentile): {decile}')
```
