```python
import random
import matplotlib.pyplot as plt

# Define the outcomes of a random variable
outcomes = [1, 2, 3, 4, 5, 6]

# Define the probabilities for each outcome (customize as needed)
probabilities = [0.166, 0.166, 0.166, 0.166, 0.166, 0.166]

# Simulate a random variable by generating random samples based on probabilities
num_samples = 1000
random_variable_samples = random.choices(outcomes, probabilities, k=num_samples)

random_variable_samples
```

```python
# Plot the histogram of the simulated random variable
plt.hist(random_variable_samples, bins=len(outcomes), align='mid', rwidth=0.8, density=True)
plt.title('Simulated Discrete Random Variable')
plt.xlabel('Outcomes')
plt.ylabel('Probability')
plt.show()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Define the outcomes of a six-sided die
outcomes = [1, 2, 3, 4, 5, 6]

# Define the probabilities for each outcome (assuming a fair die)
probabilities = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

num_samples = 100000
random_variable_samples = random.choices(outcomes, probabilities, k=num_samples)
random_variable_samples
```

```python
# Plot the probability distribution
plt.bar(outcomes, probabilities, align='center')
plt.title('Discrete Probability Distribution of a Six-sided Die')
plt.xlabel('Outcomes')
plt.ylabel('Probabilities')
plt.show()
```

```python
# Plot the histogram of the simulated random variable
plt.hist(random_variable_samples, bins=len(outcomes), align='mid', rwidth=0.8, density=True)
plt.title('Simulated Discrete Random Variable')
plt.xlabel('Outcomes')
plt.ylabel('Probability')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of possible outcomes
num_outcomes = 6

# Probability for each outcome (assuming a fair distribution)
probability = 1 / num_outcomes

# Generate random samples from a uniform distribution
uniform_samples = np.random.choice(range(1, num_outcomes + 1), size=1000, p=[probability] * num_outcomes)

# Plot the histogram
plt.hist(uniform_samples, bins=np.arange(0.5, num_outcomes + 1.5), align='mid', rwidth=0.8, density=True)
plt.title('Uniform Distribution')
plt.xlabel('Outcomes')
plt.ylabel('Probability')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of trials
num_trials = 10

# Probability of success on each trial
probability_of_success = 0.3

# Generate random samples from a binomial distribution
binomial_samples = np.random.binomial(num_trials, probability_of_success, size=10000)

binomial_samples
```

```python
# Plot the histogram
plt.hist(binomial_samples, bins=np.arange(-0.5, num_trials + 1.5), align='mid', rwidth=0.8, density=True)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Average number of events per unit interval
average_rate = 3

# Generate random samples from a Poisson distribution
poisson_samples = np.random.poisson(average_rate, size=1000)

# Plot the histogram
plt.hist(poisson_samples, bins=np.arange(-0.5, max(poisson_samples) + 1.5), align='mid', rwidth=0.8, density=True)
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of trials (for illustration purposes, set to 1 for Bernoulli)
num_trials = 1

# Probability of success on each trial
probability_of_success = 0.3

# Generate random samples from a Bernoulli distribution
bernoulli_samples = np.random.binomial(num_trials, probability_of_success, size=1000)

# Plot the histogram
plt.hist(bernoulli_samples, bins=np.arange(-0.5, num_trials + 1.5), align='mid', rwidth=0.8, density=True)
plt.title('Bernoulli Distribution')
plt.xlabel('Outcome (Success or Failure)')
plt.ylabel('Probability')
plt.xticks([0, 1], ['Failure', 'Success'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data for a normal distribution
mu = 0      # mean
sigma = 1   # standard deviation
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = norm.pdf(x, mu, sigma)

# Plot the normal distribution
plt.plot(x, y, label='Normal Distribution')
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Generate data for a t-distribution
degrees_of_freedom = 5
x = np.linspace(-3, 3, 100)
y = t.pdf(x, degrees_of_freedom)

# Plot the t-distribution
plt.plot(x, y, label=f't-Distribution (df={degrees_of_freedom})')
plt.title('t-Distribution')
plt.xlabel('t')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.show()
```
