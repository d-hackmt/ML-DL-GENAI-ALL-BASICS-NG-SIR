```python
import numpy as np

# Sample data for illustration purposes
data = np.array([1, 2, 3, 4, 5, 100])

# Min-Max Normalization (Rescaling)
min_val = np.min(data)
max_val = np.max(data)
min_max_scaled = (data - min_val) / (max_val - min_val)
print("Min-Max Normalization:", min_max_scaled)

# Z-Score Normalization (Standardization)
mean_val = np.mean(data)
std_dev = np.std(data)
z_score_scaled = (data - mean_val) / std_dev
print("Z-Score Normalization:", z_score_scaled)

# Decimal Scaling
k = np.ceil(np.log10(np.max(np.abs(data))))
decimal_scaled = data / 10**k
print("Decimal Scaling:", decimal_scaled)

# Log Transformation
log_transformed = np.log(data)
print("Log Transformation:", log_transformed)

# Unit Vector Normalization
unit_vector_scaled = data / np.linalg.norm(data)
print("Unit Vector Normalization:", unit_vector_scaled)

# Standard Scalar (Z-score normalization)
from sklearn.preprocessing import StandardScaler
standard_scalar = StandardScaler()
standard_scaled = standard_scalar.fit_transform(data.reshape(-1, 1))
print("Standard Scalar (Z-Score Normalization):", standard_scaled)

# Robust Scalar
from sklearn.preprocessing import RobustScaler
robust_scalar = RobustScaler()
robust_scaled = robust_scalar.fit_transform(data.reshape(-1, 1))
print("Robust Scalar:", robust_scaled)
```
