```python
import numpy as np

def self_information_numpy(p):
    return -np.log2(p)

# Example usage
result_self_information = self_information_numpy(1 / 64)
print(f"Self-Information: {result_self_information}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def self_information_numpy(p):
    return -np.log2(p)

# Generate a range of probabilities for plotting
probabilities = np.linspace(0.01, 1.0, 100)  # Avoiding p = 0 for log calculation

# Calculate self-information for each probability
self_information_values = self_information_numpy(probabilities)

# Plotting
plt.plot(probabilities, self_information_values, label='Self-Information')
plt.xlabel('Probability')
plt.ylabel('Self-Information (bits)')
plt.title('Self-Information vs Probability')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def entropy(probabilities):
    return -np.sum(probabilities * np.log2(probabilities))

# Example probability distribution for a six-sided die
probabilities_die = np.array([0.2, 0.1, 0.2, 0.1, 0.2, 0.2])

# Calculate entropy
entropy_die = entropy(probabilities_die)

print(f"Entropy: {entropy_die:.3f} bits")

# Plotting
labels = ['1', '2', '3', '4', '5', '6']
plt.bar(labels, probabilities_die)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Distribution of a Six-sided Die')
plt.show()
```

```python
import numpy as np

def kl_divergence(p, q):
    return np.sum(p * np.log(p / q))

# Example probability distributions
p_distribution = np.array([0.4, 0.6])
q_distribution = np.array([0.5, 0.5])

kl_value = kl_divergence(p_distribution, q_distribution)
print(f"KL Divergence: {kl_value:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def kl_divergence(p, q):
    return np.sum(p * np.log(p / q))

# Example probability distributions
p_distribution = np.array([0.4, 0.6])
q_distribution = np.array([0.5, 0.5])

# Calculate KL Divergence
kl_value = kl_divergence(p_distribution, q_distribution)
print(f"KL Divergence: {kl_value:.3f}")

# Plotting
labels = ['Event 1', 'Event 2']
plt.bar(labels, p_distribution, label='True Distribution (P)')
plt.bar(labels, q_distribution, alpha=0.5, label='Approximating Distribution (Q)')
plt.xlabel('Events')
plt.ylabel('Probability')
plt.title('KL Divergence between P and Q')
plt.legend()
plt.show()
```
